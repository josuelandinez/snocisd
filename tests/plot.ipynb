```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data_e = np.load("energy.npy")
data_c = np.load("prob.npy")
```

```python
e_hf = -120.89353160422853
e_fci = -121.47914290809689
```

```python
# plot energy
n_rep, l_eta = data_e.shape
n_rep -= 1
x = np.arange(n_rep)
y_hf = np.ones(n_rep)*e_hf
y_fci = np.ones(n_rep)*e_fci
plt.plot(x, y_hf, c='k')
plt.plot(x, y_fci, c='k')
colors = ["steelblue", 'orange', "green", "m"]
markers= [ 'd', "o", "*", "<"]
for i in range(1,l_eta):
    plt.scatter(x, data_e[1:,i], c=colors[i-1], marker=markers[i-1], label="{0:.2f}".format(data_e[0, i]))
plt.legend(loc='lower right')
plt.yticks([e_hf, e_fci], ["HF", "FCI"])
plt.savefig("singular.png", dpi=300)
```
