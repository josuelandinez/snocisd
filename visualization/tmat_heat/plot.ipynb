```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data0 =(np.load("results/hchain10_2.0_sto3g_tvecs_ndets1.npy"))
signs = np.sign(data0)
data =  np.abs(data0)
data = np.log(data)
```

```python
fig, axs = plt.subplots(3,2)
# Gridlines based on minor ticks
sns.heatmap(data[0,0,0], linewidth=1, ax=axs[0, 0])
sns.heatmap(data[1,0,0], linewidth=1, ax=axs[0, 1])
sns.heatmap(data[2,0,0], linewidth=1, ax=axs[1, 0])
sns.heatmap(data[3,0,0], linewidth=1, ax=axs[1, 1])
sns.heatmap(data[4,0,0], linewidth=1, ax=axs[2, 0])
sns.heatmap(data[5,0,0], linewidth=1, ax=axs[2, 1])
fig.savefig("spin_alpha.png", dpi=400)
```
