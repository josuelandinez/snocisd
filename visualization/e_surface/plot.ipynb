```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib
```

```python
data = np.load("results/hchain1.4_sto3g_esurf.npy")
x = data[:, 0] # 80
y = data[:, 1] # 140
z = data[:, 2]
print(x.shape)
print(y.shape)
print(z.shape)

```

```python
X = x.reshape(80,140).T
Y = y.reshape(80,140).T
Z = z.reshape(80,140).T
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
color = cm.gist_rainbow_r 
# color = "viridis"
zmin=-6.148
ax.axes.set_zlim3d(top=-6.1, bottom=zmin) 

ax.plot_surface(X, Y, Z, cmap=color , lw=0.5, rstride=1, cstride=1, alpha=1)
ax.contour(X, Y, Z, 10, linewidths=1, cmap=color , linestyles="-", offset=zmin)
# # make the grid lines transparent
# ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
# # make the panes transparent
# ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# # hide axis
# ax.set_axis_off()
```

```python
plt.contour(X, Y, Z, levels=10, linewidths=1)
```
